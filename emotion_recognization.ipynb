```python
directory = r'C:\Users\yoges\OneDrive\Desktop\mpdataset\archive\TESS Toronto emotional speech set data'
paths = []
labels = []
for dirname, _, filenames in os.walk(directory):
    for filename in filenames:
        paths.append(os.path.join(dirname, filename))
        label = filename.split('_')[-1]
        label = label.split('.')[0]
        labels.append(label.lower())
    if len(paths) == 2800:
        break
print('Dataset is Loaded')
```

```python
def waveplot(data, sr, emotion):
    plt.figure(figsize=(8,4))
    plt.title(emotion, size=20)
    librosa.display.waveshow(data, sr=sr)
    plt.show()
    
def spectogram(data, sr, emotion):
    x = librosa.stft(data)
    xdb = librosa.amplitude_to_db(abs(x))
    plt.figure(figsize=(8,4))
    plt.title(emotion, size=20)
    librosa.display.specshow(xdb, sr=sr, x_axis='time', y_axis='hz')
    plt.colorbar()
```

```python
emotion = 'angry'
path = np.array(df['speech'][df['label']==emotion])[0]
data, sampling_rate = librosa.load(path)

# Visualize waveplot
plt.figure(figsize=(8, 4))
plt.plot(data)
plt.title(f'Waveplot for {emotion} emotion')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.show()

# Visualize spectrogram
plt.figure(figsize=(8, 4))
librosa.display.specshow(librosa.amplitude_to_db(librosa.stft(data)), sr=sampling_rate, x_axis='time', y_axis='hz')
plt.colorbar(format='%+2.0f dB')
plt.title(f'Spectrogram for {emotion} emotion')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()

Audio(path)
```

## Feauture Extraction

```python
def extract_mfcc(filename):
    y, sr = librosa.load(filename, duration=3, offset=0.5)
    mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)
    return mfcc
```

```python
extract_mfcc(df['speech'][0])
```

```python
X_mfcc = df['speech'].apply(lambda x: extract_mfcc(x))
```

```python
X_mfcc
```

```python
X = [x for x in X_mfcc]
X = np.array(X)
X.shape
```

```python
## input split
X = np.expand_dims(X, -1)
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
y = enc.fit_transform(df[['label']])
```

```python
y = y.toarray()
```

```python
y.shape
```

## Create the LSTM Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

model = Sequential([
    LSTM(256, return_sequences=False, input_shape=(40,1)),
    Dropout(0.2),
    Dense(256, activation='relu'),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(7, activation='softmax')
])
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# Train the model
history = model.fit(X, y, validation_split=0.2, epochs=50, batch_size=64)
```
